```python
#import seaborn
import seaborn as sns
#Import matplot (basis of seaborn)
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
#load in a dataset
diamonds.shape
```

```python
diamonds.head()
```

```python
#to make visuals nice, down to 100 obs.
diamond = diamonds[
                diamonds.cut.isin(['Premium','Good']) &
                diamonds.color.isin(['D','F','J'])
                ].sample(n=100, random_state=22)
diamond.shape
```

```python
sns.set_style('dark') #for the aesthetic
```

```python
sns.scatterplot(x=diamond.carat, y=diamond.price)
```

```python
sns.scatterplot(x='carat', y='price',data=diamond)
```

```python
sns.scatterplot(x='x',y='y',data=diamonds)
```

```python
#seaborn allows changes in hue, marker style, and marker size
sns.scatterplot(x=diamond.carat, y=diamond.price, hue =diamond.cut,
        hue_order=['Premium','Good'],
        palette=['purple','#55CCCC']
        );
```

```python
#Continuous Variable
sns.scatterplot(x='carat',y='price',hue='depth',
                data=diamond
                );
```

```python
#s controls size
sns.scatterplot(x='carat',y='price',s=100,
                style='cut',
                style_order=['Premium','Good'],
                data=diamond
                );
```

```python
sns.scatterplot(x='carat',y='price',
                size='cut',
                size_order=['Premium','Good'],
                data=diamond,
                sizes=[150,50]
                );
```

```python
sns.scatterplot(x='carat',y='price',size='depth',data=diamond)
```

```python
#combining
sns.scatterplot(x='carat',y='price',
                hue='cut',
                style='color',
                hue_order=['Premium','Good'],
                style_order=['D','F','J'],
                data=diamond,
                s=100)
```

```python
sns.scatterplot(x='carat',y='price',
                data=diamond,
                s=100,
                alpha=0.6) #alpha adds transparency
```

```python
sns.scatterplot(x='carat',y='price',
                data=diamond,
                s=200,
                marker='*', #changes shape
                edgecolor='black') #changes edge
```
